```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
```

```python
df = pd.read_csv('CRU.csv')
```

```python
X = df.drop(['timeseries-tas-monthly-mean', 'time'], axis=1)
y = df['timeseries-tas-monthly-mean']
```

```python
rfe_dt = RFE(estimator=DecisionTreeRegressor(), n_features_to_select=0.5)
```

```python
rfe_dt.fit(X, y)
```

```python
from operator import itemgetter
features = X.columns.to_list()
for x, y in (sorted(zip(rfe_dt.ranking_ , features), key=itemgetter(0))):
    print(x, y)
```

```python
rfe_rf = RFE(estimator=RandomForestRegressor(), n_features_to_select=0.5)
```

```python
rfe_rf.fit(X, y)
```

```python
from operator import itemgetter
features = X.columns.to_list()
for x, y in (sorted(zip(rfe_rf.ranking_ , features), key=itemgetter(0))):
    print(x, y)
```
